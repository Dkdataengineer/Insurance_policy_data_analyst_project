# Generate Claims Dataset

```python
policy = pd.read_csv("D:/zopper-insurance-analysis/Data/policy_sales.csv")

display(policy)

print(policy.shape)
```

```python
# ---- CLAIMS 2025 ----

claim_days = [7,14,21,28]

eligible = policy[policy['Policy_Purchase_Date'].str[8:10].astype(int).isin(claim_days)]

claims_2025 = eligible.sample(frac=0.3)

claims_2025['Claim_Date'] = claims_2025['Policy_Start_Date']
claims_2025['Claim_Type'] = 1
```

```python
# ---- CLAIMS 2026 ----

four_year = policy[policy['Policy_Tenure']==4]

claims_2026 = four_year.sample(frac=0.1)

date_range = pd.date_range("2026-01-01","2026-02-28")

claims_2026['Claim_Date'] = np.random.choice(date_range, len(claims_2026))
claims_2026['Claim_Type'] = 2

claims = pd.concat([claims_2025,claims_2026])

claims['Claim_ID'] = range(1,len(claims)+1)
claims['Claim_Amount'] = 10000

claims = claims[['Claim_ID','Customer_ID','Vehicle_ID','Claim_Amount','Claim_Date','Claim_Type']]

claims.to_csv("D:/zopper-insurance-analysis/Data/claims.csv",index=False)
```
